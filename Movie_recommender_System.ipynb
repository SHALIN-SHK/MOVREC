```python
ratings = ratings[ratings['userId'].isin(y)]
```

```python
ratings_with_filter = ratings.merge(movies , on='movieId')
```

```python
ratings_with_filter
```

```python
num_ratings = ratings_with_filter.groupby('title')['rating'].count().reset_index()
```

```python
num_ratings
```

```python
num_ratings.rename(columns={'rating':'num_rating'},inplace=True)
```

```python
final_rating = ratings_with_filter.merge(num_ratings, on='title')
```

```python
final_rating = final_rating[final_rating['num_rating']>100]
```

```python
final_rating
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['title','userId'],inplace=True)
```

```python
final_rating.shape
```

##CF

```python
movie_pivot = final_rating.pivot_table(columns='userId' , index='title' , values='rating')
```

```python
movie_pivot
```

```python
movie_pivot.fillna(0 , inplace=True)
```

```python
movie_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
movie_sparse = csr_matrix(movie_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(movie_sparse)
```

```python
distance , suggestion = model.kneighbors(movie_pivot.iloc[200,:].values.reshape(1,-1),n_neighbors=8)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(movie_pivot.index[suggestion[i]])
```
